```python
#a. Write a function that takes an integer as an argument and returns its factorial.

def forward_factorial(n):
    product = 1
    for number in range(1, n+1):
        product = product * number
    return product

def reverse_factorial(n):
    product = 1
    for number in range(n, 0, -1):
        product = product * number
    return product

def recursive_factorial(n):
    if n == 0:
        return 1
    else: 
        return n*recursive_factorial(n-1)

print(forward_factorial(7), recursive_factorial(7), reverse_factorial(7))


```

```python
#b. Write a function that accepts a list of numbers in gallons and returns a list of numbers in liters (1 gallon = 3.78541 liters)

def gallons_to_liters(a_list):
    converter = 3.78541
    converted = []
    for number in a_list:
        number = number * converter
        converted.append(number)
    return converted

test_list = [1, 10, 100, 0.5, 5, 50]

print(gallons_to_liters(test_list))
```

```python
#c. Write a function that accepts a list of strings and keep only strings that are less than 6 characters. 

def string_keep(list_of_strings):
    keep_list = []
    for the_string in list_of_strings:
        if len(the_string) < 6:
            keep_list.append(the_string)
    return keep_list

test_list2 = ['Apple', 'Banana', 'Cherry', 'Durian', 'Eggplant', 'Fig', 'Grape']

print(string_keep(test_list2))

```

```python
#2. Functions and List of Dictionaries
# Write a function that accepts a list of dictionaries and returns the city name with the highest population. 

testlist_dict = [{"city":"San Francisco","population":884363},
{"city":"Berkeley","population":122324},
{"city":"Oakland","population":425195}]

def city_pop(list_of_dictionaries):
    highest_pop = list_of_dictionaries[0]["population"]
    highest_city = list_of_dictionaries[0]["city"]

    for dictionary in list_of_dictionaries:
        if dictionary["population"] > highest_pop:
            highest_pop = dictionary["population"]
            highest_city = dictionary["city"]

        pop_and_city = f"Population: {highest_pop} | City: {highest_city}"
        
    return pop_and_city

print(city_pop(testlist_dict))
```

```python
# 3. Function Arguments
# Write a function that returns a list of numbers in sequence. It has the following options:
# if no arguments are passed to the function, return a list of numbers from 0 to 9
# if one argument (start) is passed, return a list of numbers from start to start + 10
# if two arguments (start, count) are passed, return a list of numbers from start to start + count

def number_list(*args):
    if len(args) == 0:
        return list(range(0,10))
    elif len(args) == 1:
        return list(range(args[0], args[0]+10))
    elif len(args) == 2:
        return list(range(args[0], args[0] + args[1]))

print(number_list(), number_list(5), number_list(3,5))
```

```python
# 4. (Optional) Write a function that accepts a string and an optional number (n) as arguments. 
# It should return:
# a string in all uppercase if only the string argument is passed to it
# a string in all uppercase if n is 0
# if n > 0, a string with the first n characters as uppercase and the rest of the string in its original case

def func(a_string, n=None):
    if n == None or n == 0:
        return a_string.upper()
    elif n > 0:
        return a_string[:n].upper() + a_string[n:]


print(func("miguel"))
print(func("miguel", 0))
print(func("miguel", 4))
```
