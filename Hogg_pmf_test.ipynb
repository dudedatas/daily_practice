```python
import matplotlib.pyplot as plt

# Define a pmf from exercise 1.6.1

def pmf(x):
    if x == 0:
        return 1 / 6
    elif x in range(1,6):
        return (12-2*x) / 36
    
    else: return 0

x_vals = [x for x in range(-2,6)]
y_vals = [round(pmf(x), 4) for x in range(-2,6)]


plt.bar(x_vals, y_vals, color="green")
plt.show()

```

```python
# Define a pmf from exercise 1.6.2

def pmf_(x):
    def fact(n):
        if n == 0:
            return 1
        return n*fact(n-1)
    
    if x in range(0,5):
        return 4*3*2*1*(1/2)**4 / (fact(x) * fact(4-x))
    else:
        return 0

x1 = [x for x in range(-2,7)]
y1 = [round(pmf_(x), 4) for x in range(-2,7)]


plt.bar(x1, y1, color="darkblue")
plt.show()
```
