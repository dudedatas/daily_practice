```python
# 1.5.8 Hogg_McKean CDF visualizer 
# Book is showing R-code but I am trying it in Python without using Numpy

import matplotlib.pyplot as plt

#Define the CDF 
def F(x):
    if x < -1:
        return 0
    elif -1 <= x < 1:
        return (x + 2) / 4 
    else:
        return 1
    
#Create linspace manually instead of using numpy.linspace to create a smooth range of numbers

x_vals = []
y_vals = []

x = -2
step = 0.01

while x < 2:
    x_vals.append(x)
    y_vals.append(F(x))
    x += step

#plot
plt.plot(x_vals, y_vals, color="darkblue")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("CDF of X")
plt.grid(True)

plt.scatter([-1,1], [F(-1), F(1)], color="red")
plt.axvline(x=-1, linestyle="--", color="darkgray")
plt.axvline(x=1, linestyle="--", color="darkgray")


plt.show()
```
