```python
# Faster than recursive but ugle

def c(n,k):

    k = min(k, n-k)

    numerator = 1
    denominator = 1
    
    for i in range(n, (n-k), -1):
        numerator *= i
    for j in range(k, 0, -1):
        denominator *= j
    
    return numerator // denominator


c(10,2)
```

```python
# Choosefunc is Slower than iteration style

def choose(n,k):
    if k == 0 or k ==n:
        return 1
    return (choose(n-1,k-1) + choose(n-1, k))

def binomial_pmf(n, k, p):
    return c(n,k)*p**k*(1-p)**(n-k)

def hyper(N, k, n, x):
    if x in range(0,n+1) and x <= k:
        return c(k,x) * c((N-k), (n-x)) / c(N, n)
    
    else : return 0



hyper(100, 30, 10, 1)

```
