```python
# 1. Higher Order Functions

# a. Using the map() function, write code to convert a list of numbers in gallons to liters (1 gallon = 3.78541 liters)


mylist = [1, 10, 100, 0.5, 5, 50]

def converter(n):
    return n * 3.78541

g2l = list(map(converter, mylist))

print(g2l)

```

```python
# b. Using the filter() function, write code that accepts a list of strings and keep only strings that are less than 6 characters. 

def string_keeper(StringList):
    return len(StringList) < 6

s_list = ['Apple', 'Banana', 'Cherry', 'Durian', 'Eggplant', 'Fig', 'Grape']

values = filter(string_keeper, s_list)

print(list(values))
```
